```python
import json
import pandas as pd
import numpy as np
import os
# from IPython.display import display, HTML 
import csv
import json

from tweepy.streaming import StreamListener
from tweepy import OAuthHandler, API, Cursor


# Load Twitter Streaming API credentials for authentication
with open("twitter_credentials.json", "r") as cred_file: 
    creds = json.load(cred_file)
        
# authenticate        
auth = OAuthHandler(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])
auth.set_access_token(creds['ACCESS_TOKEN'], creds['ACCESS_SECRET'])
api = API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)


# Create an error log file
errorlog = open("errorlog.txt","w") 
errorlog.write("")
errorlog.close() 


datadir = os.getcwd()+ '\\Tweets_2019\\ids'
entries = os.listdir(datadir)
for entry in entries:
    if entry.endswith(".txt"):
#         print(os.getcwd())
        curfile = os.getcwd()+ '\\Tweets_2019\\ids\\' + entry
        print(curfile)
        file = open(curfile, 'r')
        lines = file.readlines()
        file.close
        
        # destination file
        filename = entry[:len(entry) - 4]
        csv_dest_file = os.getcwd()+ '\\tweets\\' + filename + '.csv'
        json_dest_file = os.getcwd()+ '\\tweets\\' + filename + '.txt'
        print(csv_dest_file)
        
        # Create csv file to store all tweets of the hour in
        with open(csv_dest_file,mode='w', newline='') as fd:
            tweet_writer = csv.writer(fd)
            tweet_writer.writerow(['tweetid', 'in_reply_to_status_id',
                      'in_reply_to_user_id', 'user.screen_name',
                      'user.id', 'user.verified', 'user.description',
                     'user.friends_count', 'user.followers_count',
                     'created_at', 'full_text', 'retweet_count',
                     'favorite_count'])
        
        # clear content of tweet destination file
        open(json_dest_file, 'w').close()    
        
        
        for line in lines:
            tweet_id = line.strip('\n')
#             print(tweet_id)
            
            try:
                tweet = api.get_status(int(tweet_id), tweet_mode="extended")
                tweetid = int(tweet_id)
                
                in_reply_to_status_id = tweet.in_reply_to_status_id
                in_reply_to_user_id = tweet.in_reply_to_user_id
                user_screen_name = tweet.user.screen_name
                user_id = tweet.user.id
                user_verified = tweet.user.verified
                user_description = tweet.user.description
                user_friends_count = tweet.user.friends_count
                user_followers_count = tweet.user.followers_count
                created_at = tweet.created_at
                full_text = tweet.full_text
                retweet_count = tweet.retweet_count
                favorite_count = tweet.favorite_count
#                 quote_count = tweet.quote_count
#                 entities_hashtags = "".join(tweet.entities.hashtags)
#                 entities_urls = "".join(tweet.entities.urls)
#                 entities_user_mentions = "".join(tweet.entities.user_mentions)

#                 print(in_reply_to_status_id)
#                 print(in_reply_to_user_id)
#                 print(user_screen_name)
#                 print(user_id)
#                 print(user_verified)
#                 print(user_description)
#                 print(user_friends_count)
#                 print(user_followers_count)
#                 print(created_at)
#                 print(full_text)
#                 print(retweet_count)
#                 print(favorite_count)
    
                # Write tweet content to csv file
                with open(csv_dest_file,mode='a', newline='') as fd:
                    tweet_writer = csv.writer(fd)
                    tweet_writer.writerow([tweet_id, in_reply_to_status_id, in_reply_to_user_id, user_screen_name, user_id, 
                             user_verified, user_description, user_friends_count, user_followers_count,
                             created_at,
                              full_text,
                              retweet_count,
                              favorite_count])
                    
                # Write complete tweet to JSON file
                with open(json_dest_file, 'a') as outfile:
#                     json_str = json.dumps(tweet._json)
#                     print(json_str)
                    # Sort by key? 
                    json.dump(tweet._json, outfile, indent=4)
                    
                
                

            except Exception as e:
#                 print(e.reason)
                errorlog = open("errorlog.txt","a")
                errorlog.write(e.reason)
                errorlog.write('\n')
                errorlog.close() 

                pass    
        
    else:
        continue






```
